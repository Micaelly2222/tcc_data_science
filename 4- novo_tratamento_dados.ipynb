```
from unidecode import unidecode
import joblib
from pathlib import Path
import pandas as pd
import numpy as np
from orjson import dumps
```

```
df_merged3 = joblib.load('df_merged2.joblib')
df_merged3
```

```
dados = pd.read_excel('dados2.xlsx', index_col=0, usecols='A:H')
dados
```

```
# e padronizei os indices : deixei tudo maisculo e sem acento 
dados.index = dados.index.map(unidecode)
dados.index = dados.index.str.upper()
dados
```

```
dados.info()
```

```
# criei uma estrutura de colunas em formato de MultiIndex para o dataFrame "dados"
# cada coluna terá três níveis: (ibge), o ano e a variável

nomes = [('ibge', '2017', 'IDEB – Anos iniciais do ensino fundamental (Rede pública)'),
              ('ibge', '2019', 'IDEB – Anos iniciais do ensino fundamental (Rede pública)'),
              ('ibge', '2017', 'IDEB – Anos finais do ensino fundamental (Rede pública)'),
              ('ibge', '2019', 'IDEB – Anos finais do ensino fundamental (Rede pública)'),
              ('ibge', '2017', 'PIB per capita'),
              ('ibge', '2018', 'PIB per capita'),
              ('ibge', '2019', 'PIB per capita')]
dados.columns = pd.MultiIndex.from_tuples(nomes)
dados
```

```
dados.info()
```

```
def trata_dados(elemento):
    if elemento == "-":
        return np.nan
    if isinstance(elemento, str):
        return pd.to_numeric(elemento)
    if isinstance(elemento, (float, int)):
        return elemento 
    raise ValueError( f"Tipo {type(elemento)} não esperado")
```

```
dados2 = dados.applymap(trata_dados)
dados2
```

```
dados2.info()
```

```
df_merged4 = df_merged3.join(dados2, how = 'outer')
df_merged4
```

```
# salvei em um arquivo joblib
# Também salvei os dados em um arquivo CSV e os tipos em um json
# fiz isso para poder compartilhar o arquivo em formato csv

dtypes = df_merged4.dtypes.apply(lambda x: x.name).to_dict()
with open('df_merged4_dtypes.json', 'wb') as fp:
    fp.write(dumps({','.join(k): v for k, v in dtypes.items()}))
df_merged4.to_csv('df_merged4.gz')
joblib.dump(df_merged4, 'df_merged4.joblib')
```
