```python
# automatizei a extração de dados do site do TCE/MT
# coletei os dados de todos os municípios, relativos aos limites da LRF, de 2017 a 2022
# utilizei a biblioteca Selenium para o web scrapping
# salvei os dados em arquivos .h5. para depois manipular os dados no jupyter usando o Pandas 
# arquivo do Selenium localizado em C:\Users\Pichau\PycharmProjects\pythonProject10\Outs


# utilizei a biblioteca joblib para carregar os dados salvos em arquivos .h5
# aqui estou verificando quais index e colunas do arquivo


import joblib
from pathlib import Path
import pandas as pd

dir_path = Path(r'C:\Users\Pichau\PycharmProjects\pythonProject10\Outs')
file_list = set(dir_path.glob('*'))

indexes = set()
columns = set()

for file_path in file_list:
    data = joblib.load(file_path)
    if isinstance(data, pd.DataFrame):
        indexes.update(data.index.unique())
        columns.update(data.columns)
    elif isinstance(data, list):
        for df in data:
            indexes.update(df.index.unique())
            columns.update(df.columns)

print("Index:")
for index in indexes:
    print(index)

print("Colunas:")
for column in columns:
    print(column)
```

```python
# para entender como os dados estao estruturados, verifiquei o modelo do arquivo com base nas colunas e index


import joblib
from pathlib import Path
import pandas as pd
from IPython.display import display

dir_path = Path(r'C:\Users\Pichau\PycharmProjects\pythonProject10\Outs')
file_list = set(dir_path.glob('*'))


set_modelos = set()

for file_path in file_list:
    data = joblib.load(file_path)
    if isinstance(data, pd.DataFrame):
        modelo = (data.index.name, data.columns.name, tuple(data.index), tuple(data.columns))
        if modelo not in set_modelos:
            # Se não estiver, exibe e salva no set o novo modelo
            set_modelos.add(modelo)
            display(data.head())
    elif isinstance(data, list):
        for df in data:
            modelo = (df.index.name, df.columns.name, tuple(df.index), tuple(df.columns))
            if modelo not in set_modelos:
                set_modelos.add(modelo)
                display(df.head())
```

```python
# fiz uma funcao trazendo as informaçoes que preciso, baseada no index e colunas escolhidas
# fiz um for para percorrer as informaçoes para cada municipio e ano
# criei o dataframe( chamado "df")

import joblib
from pathlib import Path
import pandas as pd
from collections import defaultdict

dir_path = Path(r'C:\Users\Pichau\PycharmProjects\pythonProject10\Outs')
file_list = set(dir_path.glob('*'))

dict_final_df = defaultdict(dict)

def final_df(df_dict, region, year, df):
        if df.index.name == 'DESPESA COM PESSOAL DO PODER EXECUTIVO':
            df_dict[region]['EXECUTIVO', year,'total_despesas_com_pessoal'] = df.loc['Total despesa com pessoal (R$)', 'VALORES']
            df_dict[region]['EXECUTIVO', year,'rcl_acumulado'] = df.loc['RCL Acumulado (R$)', 'VALORES']
            df_dict[region]['EXECUTIVO', year,'aplicado'] = df.loc['Aplicado (%)', 'VALORES']
            df_dict[region]['EXECUTIVO', year,'limite_legal'] = df.loc['Limite Legal (%)', 'VALORES']
            df_dict[region]['EXECUTIVO', year,'alerta_90'] = df.loc['Alerta 90%', 'VALORES']
        elif df.index.name == 'DESPESA COM PESSOAL DO PODER LEGISLATIVO':
                df_dict[region]['LEGISLATIVO', year,'despesas'] = df.loc['Total despesa com pessoal (R$)', 'VALORES']
                df_dict[region]['LEGISLATIVO', year,'rcl_acumulado'] = df.loc['RCL Acumulado (R$)', 'VALORES']
                df_dict[region]['LEGISLATIVO', year,'aplicado'] = df.loc['Aplicado (%)', 'VALORES']
                df_dict[region]['LEGISLATIVO', year,'limite_legal'] = df.loc['Limite Legal (%)', 'VALORES']
                df_dict[region]['LEGISLATIVO', year,'alerta_90'] = df.loc['Alerta 90%', 'VALORES']
        elif df.index.name ==  'DÍVIDA CONSOLIDADA LÍQUIDA':      
                df_dict[region]['DÍVIDA', year, 'rcl'] = df.loc['Receita Corrente Líquida', 'VALORES']
                df_dict[region]['DÍVIDA', year, 'limite_maximo_120'] = df.loc['Limite máximo de 120% da RCL para o total da Dívida Consolidada Líquida', 'VALORES']
                df_dict[region]['DÍVIDA', year, 'limite_maximo_90'] = df.loc['Limite de 90% do limite máximo para fins de alerta', 'VALORES']
                df_dict[region]['DÍVIDA', year, 'total_divida'] = df.loc['Total da Dívida Consolidada Líquida', 'VALORES']
                df_dict[region]['DÍVIDA', year, '%_limite_maximo'] = df.loc['% do limite máximo da Dívida Consolidada Líquida', 'VALORES']                                                      
        elif  df.index.name == 'OPERAÇÃO DE CRÉDITO':
                df_dict[region]['CRÉDITO', year, 'rcl'] = df.loc['Receita Corrente Líquida', 'VALORES']
                df_dict[region]['CRÉDITO', year, 'limite_maximo_16'] = df.loc['Limite máximo de 16% da RCL para o total das operações de crédito', 'VALORES']
                df_dict[region]['CRÉDITO', year, 'limite_maximo_90'] = df.loc['Limite de 90% do limite máximo para fins de alerta', 'VALORES']
                df_dict[region]['CRÉDITO', year, 'total_operacoes'] = df.loc['Total das operações de crédito', 'VALORES']
                df_dict[region]['CRÉDITO', year, '%_limite_maximo'] = df.loc['% do limite máximo das operações de crédito', 'VALORES']                                                       
        elif df.index.name == 'CONCESSÃO DE GARANTIAS':
                df_dict[region]['CONCESSÃO', year, 'rcl'] = df.loc['Receita Corrente Líquida', 'VALORES']
                df_dict[region]['CONCESSÃO', year, 'limite_maximo_22'] = df.loc['Limite máximo de 22% da RCL para concessão de garantia', 'VALORES']
                df_dict[region]['CONCESSÃO', year, 'limite_maximo_90'] = df.loc['Limite de 90% do limite máximo para fins de alerta', 'VALORES']
                df_dict[region]['CONCESSÃO', year, 'total_garantias'] = df.loc['Total de Garantias Concedidas', 'VALORES']
                df_dict[region]['CONCESSÃO', year, '%_limite_maximo'] = df.loc['% do limite máximo de garantias concedidas', 'VALORES']                                             
        elif df.index.name == 'RESULTADO PRIMÁRIO':
                df_dict[region]['RESULTADO', year, 'receitas_fiscais'] = df.loc['Receitas fiscais', 'VALORES']
                df_dict[region]['RESULTADO', year, 'despesas_fiscais'] = df.loc['Despesas fiscais', 'VALORES']
                df_dict[region]['RESULTADO', year, 'resultado_primario'] = df.loc['Resultado Primário', 'VALORES']
                df_dict[region]['RESULTADO', year, 'meta_resultado'] = df.loc['Meta de resultado primário fixada no Anexo de Metas Fiscais da LDO', 'VALORES']                                                          

# fiz um for que passa sobre a lista de arquivos e carrega os dados de cada arquivo usando joblib
# ele analisa o nome do arquivo para obter o municipio e o ano correspondente            
                
                
for file_path in file_list:
    data = joblib.load(file_path)
    file_name = file_path.stem
    region, year = file_name.rsplit('-', maxsplit=1)
    if isinstance(data, pd.DataFrame):
        final_df(dict_final_df, region, year, data)
    elif isinstance(data, list):
        for df in data:
            if isinstance(df, pd.DataFrame):
                final_df(dict_final_df, region, year, df)
                
# criando dataframe                
df = pd.DataFrame(dict_final_df)
df = df.T
df.head()
```

```python
# busquei mais variaveis sobre os municipios (que estao em dados.xlsx), dados do IBGE
# aqui estou trazendo os dados usando o pandas
# contém informações adicionais sobre os municípios

import pandas as pd
dados = pd.read_excel(r'C:\Users\Pichau\Desktop\TCC MBA\dados.xlsx')
dados
```

```python
# setando municipio do dataframe "dados" como index
# e padronizei os indices : deixei tudo maisculo e sem acento e fiz o mesmo pro df anterior
dados.set_index('Municipios', inplace=True)
dados.index = dados.index.str.upper()
dados.index = dados.index.map(unidecode)
df.index = df.index.map(unidecode)
```

```python
dados
```

```python
# criei uma estrutura de colunas em formato de MultiIndex para o dataFrame "dados"
# cada coluna terá três níveis: (ibge), o ano (2020) e a variável

nomes = [('ibge', 2020, 'População no último censo'),
              ('ibge', 2020, 'Salário médio mensal dos trabalhadores formais'),
              ('ibge', 2020, 'Índice de Desenvolvimento Humano Municipal (IDHM)'),
              ('ibge', 2020, 'PIB per capita'),
              ('ibge', 2020, 'IDEB – Anos iniciais do ensino fundamental (Rede pública)'),
              ('ibge', 2020, 'IDEB – Anos finais do ensino fundamental (Rede pública)'),
              ('ibge', 2020, 'Área da unidade territorial'),
              ('ibge', 2020, 'Mesorregião')]
dados.columns = pd.MultiIndex.from_tuples(nomes)
```

```python
# aqui uni os dois dataframes: o "df" (dados do TCE) com o "dados" (dados do IBGE)
# usei "outer" para todos os registros ser incluidos
# o resultado disso é o "df_merged"

df_merged = df.join(dados, how = 'outer')
df_merged
```

```python
joblib.dump(df_merged, 'df_merged.joblib')
```
